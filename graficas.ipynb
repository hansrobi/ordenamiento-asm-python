```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funci√≥n para leer datos de un archivo
def leer_datos(nombre_archivo):
    datos = np.loadtxt(nombre_archivo)
    return datos[:, 0], datos[:, 1], datos[:, 2], datos[:, 3], datos[:, 4], datos[:, 5], datos[:, 6]
```

```python
# Leer los datos de ambos archivos
tamanos_asm, bs_asc_asm, qs_asc_asm, bs_desc_asm, qs_desc_asm, bs_rand_asm, qs_rand_asm = leer_datos('tiempos_asm.txt')
tamanos_py, bs_asc_py, qs_asc_py, bs_desc_py, qs_desc_py, bs_rand_py, qs_rand_py = leer_datos('tiempos_python.txt')
```

```python
# Crear la figura con 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 16))

# Funci√≥n para graficar en un subplot
def graficar(ax, tamanos_asm, bs_asm, qs_asm, tamanos_py, bs_py, qs_py, titulo):
    ax.plot(tamanos_asm, bs_asm, label='Bubble Sort (ASM)', marker='o')
    ax.plot(tamanos_asm, qs_asm, label='Quick Sort (ASM)', marker='s')
    ax.plot(tamanos_py, bs_py, label='Bubble Sort (Python)', marker='^')
    ax.plot(tamanos_py, qs_py, label='Quick Sort (Python)', marker='D')
    ax.set_xlabel('Tama√±o del arreglo')
    ax.set_ylabel('Tiempo (segundos)')
    ax.set_title(titulo)
    ax.legend()
    ax.grid(True)

# Graficar para arreglos ascendentes
graficar(ax1, tamanos_asm, bs_asc_asm, qs_asc_asm, tamanos_py, bs_asc_py, qs_asc_py, 'Tiempos de ejecuci√≥n - Orden Ascendente')

# Graficar para arreglos descendentes
graficar(ax2, tamanos_asm, bs_desc_asm, qs_desc_asm, tamanos_py, bs_desc_py, qs_desc_py, 'Tiempos de ejecuci√≥n - Orden Descendente')

# Graficar para arreglos aleatorios
graficar(ax3, tamanos_asm, bs_rand_asm, qs_rand_asm, tamanos_py, bs_rand_py, qs_rand_py, 'Tiempos de ejecuci√≥n - Orden Aleatorio')

plt.tight_layout(pad=3.0)
plt.show()
```
